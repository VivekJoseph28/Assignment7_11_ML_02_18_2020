In this assignment students have to find the frequency of words in a
webpage. User can use urllib and BeautifulSoup to extract text from
webpage.

### Importing the packages

```python
from bs4 import BeautifulSoup 
import urllib.request 
import nltk 
from nltk.corpus import stopwords 
nltk.download("stopwords")
```

### Reading the Html pages using beautiful soup and text splitting by defining tokens and stopwords

```python
response = urllib.request.urlopen('http://en.wikipedia.org/wiki/Valley_of_Flowers_National_Park') 
html = response.read() 
soup = BeautifulSoup(html,"html5lib") 
text = soup.get_text(strip=True) 
tokens = [t for t in text.split()] 
clean_tokens = tokens[:] 
sr = stopwords.words('english') 
for token in tokens: 
    if token in stopwords.words('english'): 
        clean_tokens.remove(token) 
freq = nltk.FreqDist(clean_tokens) 
for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```

### Frequency Plot of Words Distribution in the Html Page

```python
freq.plot(20,cumulative=False)
```
